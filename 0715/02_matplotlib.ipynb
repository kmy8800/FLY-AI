```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.random.randn(150)
```

```python
plt.plot(x)
plt.show()
```

<img src="./plt.png" width="400"/>

```python
plt.figure()
plt.subplot(1, 2, 1)
plt.plot(x)
plt.subplot(1, 2, 2)
plt.plot(x)
plt.show()
```

```python
fig = plt.figure(2)
ax1 = fig.add_subplot(1, 2, 1)
ax1.plot(x)
ax2 = fig.add_subplot(1, 2, 2)
ax2.plot(x)
plt.show()
```

```python
fig, axes = plt.subplots(2, 2)
axes = axes.flatten()
for i in range(4):
    axes[i].plot(x)
```

```python
x = np.random.rand(15).cumsum()
y = np.random.rand(15).cumsum()
z = np.random.rand(15).cumsum()
```

```python
plt.figure(figsize=(8, 6))
plt.plot(x, label="x", color='k', linestyle=":", marker='x')
plt.plot(y, label="y")
plt.plot(z, label="z")
plt.title("Title")
plt.xlabel("xlabel")
plt.ylabel("ylabel")
plt.legend()
plt.show()
```

```python
plt.figure(3, figsize=(8,6))
plt.scatter(x, y, marker='*', label='x', c='gold', s=10000)
plt.scatter(y, x, marker='o', label='o')

plt.legend()
plt.show()
```

```python
year = ['A', 'B', 'C']
points = [100, 200, 300]
plt.bar(year, points, color=['r', 'g', 'b'])  # plt.barh(year, points)
```

```python
x = np.random.normal(0, 1, 1000000)
```

```python
plt.figure(4)
plt.hist(x, bins=1000, color='pink')
plt.show()
```

```python
plt.boxplot([x, y, z])
plt.show()
```
