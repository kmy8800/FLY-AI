# Seaborn

[seaborn](https://seaborn.pydata.org/)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic = pd.read_csv('./dataset/titanic_train.csv')
```

```python
titanic.head()
```

```python
plt.figure(figsize=(8,6))
sns.histplot(data=titanic, x='Age')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.histplot(data=titanic, x='Age', bins=8)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.histplot(data=titanic, x='Age', bins=8, kde=True)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.histplot(data=titanic, x='Age', bins=8, hue='Survived')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.histplot(data=titanic, x='Age', bins=8, hue='Sex')
plt.show()
```

## displot
* figure-level plot
- matplotlib과 연결되지 않음(따로 설정 필요)

```python
sns.displot(data=titanic, x='Age', kind='kde')
plt.show()
```

```python
sns.displot(data=titanic, x='Age', col='Survived')
plt.show()
```

```python
sns.displot(data=titanic, x='Age', col='Survived', hue='Sex')
plt.show()
```

## bar plot

```python
sns.countplot(data=titanic, x='Pclass')   
```

```python
sns.barplot(data=titanic, x='Pclass', y='Survived', ci=False)
```

```python
sns.barplot(x='Pclass', y='Fare', data=titanic, ci=False, hue='Sex')
```

## Point plot

```python
sns.pointplot(data=titanic, x='Pclass', y='Fare', hue='Sex')
plt.show()
```

## Box plot

```python
sns.boxplot(data=titanic, x='Pclass', y='Age', hue='Sex')
plt.show()
```

## Violin plot

```python
sns.violinplot(data=titanic, x='Pclass', y='Age', hue='Sex', split=True)
plt.show()
```

## Count plot

```python
sns.countplot(data=titanic, x='Embarked')
plt.show()
```

## Heatmap

```python
arr1 = np.random.rand(4, 5)
```

```python
sns.heatmap(data=arr1, annot=True, cmap='YlGnBu')
```

## Scatter plot

```python
tips = sns.load_dataset('tips')
sns.scatterplot(x='total_bill', y='tip', data=tips, hue='sex')
plt.show()
```

```python
sns.regplot(data=tips, x='total_bill', y='tip')
plt.show()
```

## Pair plot

```python
sns.set_theme(style='whitegrid')
pengsu = sns.load_dataset('penguins')
sns.pairplot(data=pengsu, hue='species')
plt.show()
```
